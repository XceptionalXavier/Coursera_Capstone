```python
# import libraries for website scraping
from bs4 import BeautifulSoup 
import urllib.request # for parsing requests
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read() # reading the file
soup = BeautifulSoup(source,'lxml') #parsing the page to soup
```

```python
# reading the file to pandas
table = soup.find_all('table')[0] 
table_rows = table.find_all('tr')
```

```python
# converting the read file into pandas dataframe
dataList = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    if "Not assigned" not in row: # removing cells with non assigned
        dataList.append(row)
dataFrame = pd.DataFrame(dataList, columns=['Postal code', 'Borough', 'Neighborhood'])
dataFrame
dataFrame.drop([0]) #dropping the first row with none values
```

```python
dataFrame.Neighborhood = pd.Index(map(lambda x : str(x)[:-1], dataFrame.Neighborhood)) # to get away the \n
```

```python
dataFrame.groupby('Borough')['Neighborhood'].apply(', '.join).reset_index() # Combining the multiple cells into 1
```
